**Removing outliers**

```python

# IQR Method for removing outliers
Q1 = df6['price'].quantile(0.25)
Q3 = df6['price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Removing outliers
df_cleaned = df6[(df6['price'] >= lower_bound) & (df6['price'] <= upper_bound)]

df_cleaned
```
