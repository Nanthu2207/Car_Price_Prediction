**Bangalore**

```python
data_1=pd.DataFrame(pd.read_excel("/content/bangalore_cars.xlsx"))
data_1
```

```python
import pandas as pd
import ast
b1=[]
for i in data_1["new_car_detail"]:
 b1.append(ast.literal_eval(i))

# Assuming 'da_d' is a list of dictionaries
df_d= pd.DataFrame(b1)

# Extract the nested dictionary 'trendingText' and create new columns
df_d[['trending_imgUrl', 'trending_heading', 'trending_desc']] = df_d['trendingText'].apply(pd.Series)

# Remove the original 'trendingText' column
df1 = df_d.drop('trendingText', axis=1)


# Assuming data_1["new_car_overview"] contains string representations of dictionaries
b2= [ast.literal_eval(i) for i in data_1["new_car_overview"]]

# Create DataFrame from the list of dictionaries
df_o= pd.DataFrame(b2)

# Function to extract 'value' for a specific key in 'top'
def extract_value(top_data, key_name):
    for item in top_data:
        if item['key'] == key_name:
            return item['value']
    return None


# Create new columns by applying the function to extract values
df_o['Registration_Year'] = df_o['top'].apply(lambda x: extract_value(x, 'Registration Year'))
df_o['Insurance_Validity'] = df_o['top'].apply(lambda x: extract_value(x, 'Insurance Validity'))
df_o['Fuel_Type'] = df_o['top'].apply(lambda x: extract_value(x, 'Fuel Type'))
df_o['Seats'] = df_o['top'].apply(lambda x: extract_value(x, 'Seats'))
df_o['Kms_Driven'] = df_o['top'].apply(lambda x: extract_value(x, 'Kms Driven'))
df_o['RTO'] = df_o['top'].apply(lambda x: extract_value(x, 'RTO'))
df_o['Ownership'] = df_o['top'].apply(lambda x: extract_value(x, 'Ownership'))
df_o['Engine Displacement'] = df_o['top'].apply(lambda x: extract_value(x, 'Engine Displacement'))
df_o['Transmission'] = df_o['top'].apply(lambda x: extract_value(x, 'Transmission'))
df_o['Year_of_Manufacture'] = df_o['top'].apply(lambda x: extract_value(x, 'Year of Manufacture'))

# Drop the original 'top' column
df_o.drop('top', axis=1, inplace=True)


df_new_car_feature = data_1['new_car_feature']

# Initialize an empty list to store all the processed rows
processed_features = []

# Iterate over the 'new_car_feature' column in the DataFrame
for feature_str in df_new_car_feature:
    # Parse the string to a dictionary using ast.literal_eval
    data = ast.literal_eval(feature_str)

    # Helper function to extract values from a list of dictionaries
    def extract_values(dict_list):
         return [item['value'] for item in dict_list]


    # Organize data into the required categories with index checking
    top = extract_values(data['top'])
    comfort = extract_values(data['data'][0]['list']) if len(data['data']) > 0 else []
    interior = extract_values(data['data'][1]['list']) if len(data['data']) > 1 else []
    exterior = extract_values(data['data'][2]['list']) if len(data['data']) > 2 else []
    safety = extract_values(data['data'][3]['list']) if len(data['data']) > 3 else []
    entertainment = extract_values(data['data'][4]['list']) if len(data['data']) > 4 else []

    # Call the extract_values function with the 'commonIcon' data from the dictionary,
    # handling cases where 'commonIcon' might be missing.
    commonIcon = data['commonIcon']

    # Create a single-row DataFrame with 6 columns
    processed_features.append({
        'top': ', '.join(top),
        'comfort': ', '.join(comfort),
        'interior': ', '.join(interior),
        'exterior': ', '.join(exterior),
        'safety': ', '.join(safety),
        'entertainment': ', '.join(entertainment),
        'commonIcon': ''.join(commonIcon)
    })



# Assuming data_1["new_car_specs"] contains string representations of dictionaries
b3= [ast.literal_eval(i) for i in data_1["new_car_specs"]]

# Create DataFrame from the list of dictionaries
df_s= pd.DataFrame(b3)

def extract_value(data, key_name, from_list=False):
    """ Extract 'value' for a specific key from a list or dictionary of dictionaries. """
    if isinstance(data, list):
        for item in data:
            if from_list and 'list' in item:
                for inner_item in item['list']:
                    if inner_item['key'] == key_name:
                        return inner_item['value']
            elif item['key'] == key_name:
                return item['value']
    return None

# Define the columns and their keys
columns_top = {
    'Mileage': 'Mileage',
    'Engine': 'Engine',
    'Max_Power': 'Max Power',
    'Torque': 'Torque',
    'Wheel_Size': 'Wheel Size',
    'Seats': 'Seats'
}

columns_data = {
    'Color': 'Color',
    'Engine_Type': 'Engine Type',
    'Displacement': 'Displacement',
    'Max_Power': 'Max Power',
    'Max_Torque': 'Max Torque',
    'No_of_Cylinder': 'No of Cylinder',
    'Values_per_Cylinder': 'Values per Cylinder',
    'Value_Configuration': 'Value Configuration',
    'Fuel_Supply_System': 'Fuel Supply System',
    'Turbo_Charger': 'Turbo Charger',
    'Length': 'Length',
    'Width': 'Width',
    'Height': 'Height',
    'Wheel_Base': 'Wheel Base',
    'Kerb_Weight': 'Kerb Weight',
    'Gear_Box': 'Gear Box',
    'Seating_Capacity': 'Seating Capacity',
    'Steering_Type': 'Steering Type',
    'Front_Brake_Type': 'Front Brake Type',
    'Rear_Brake_Type': 'Rear Brake Type',
    'Tyre_Type': 'Tyre Type',
    'Alloy_Wheel_Size': 'Alloy Wheel Size',
    'No_Door_Numbers': 'No Door Numbers'
}

# Create new columns for 'top'
for col, key in columns_top.items():
    df_s[col] = df_s['top'].apply(lambda x: extract_value(x, key))

# Create new columns for 'data'
for col, key in columns_data.items():
    df_s[col] = df_s['data'].apply(lambda x: extract_value(x, key, from_list=True))

# Drop the original 'top' and 'data' columns
df_s.drop(['top', 'data'], axis=1, inplace=True)


import ast
b5=[]
for i in data_1["car_links"]:
    b5.append(i)


car_detail=pd.DataFrame(df1)
car_overview=pd.DataFrame(df_o)
car_features=pd.DataFrame(processed_features)
car_specs=pd.DataFrame(df_s)
car_links=pd.DataFrame(b5)
df1['City']='Bangalore'

df_combined_1=pd.concat([car_detail,car_overview,car_features,car_specs,car_links,df1['City']],axis=1)
df_combined_1
```
