**Model Training**

```python
x = df[['Kms_Driven','Year_of_Manufacture','Mileage','Ownership','Transmission','Insurance_Type','Fuel_Type','Body_Type','City',]].values
y = df['price'].values
```

```python
# train-test split
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

# Model Performance

**Linear regression**

```python
# import linear regression
from sklearn.linear_model import LinearRegression
model=LinearRegression().fit(X_train,y_train)
model

y_pred=model.predict(X_test)
y_pred
```

```python
from sklearn import metrics
import numpy as np
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print('R² score:', r2) #model accuracy

```

**Decision Tree**

```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeRegressor(random_state=10)
clf.fit(X_train,y_train)
y_train_pred = clf.predict(X_train)
y_test_pred = clf.predict(X_test)
```

```python
from sklearn import metrics
import numpy as np
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_test_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_test_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_test_pred)))

from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_test_pred)
print('R² score:', r2)
```

**Gradient Boosting**

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
cf= GradientBoostingRegressor(random_state=10,n_estimators=10)
cf.fit(X_train,y_train)
y_train_pred = cf.predict(X_train)
y_test_pred = cf.predict(X_test)
```

```python
from sklearn import metrics
import numpy as np
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_test_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_test_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_test_pred)))

from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_test_pred)
print('R² score:', r2)
```

**Random Forest**

**Best Hyperparameters**

```python
from sklearn.model_selection import GridSearchCV
params = {'max_depth': [4,6,8,10,12],
        'min_samples_split': [2,3,4],
        'min_samples_leaf': [1,2]}


clf = RandomForestRegressor()
gcv = GridSearchCV(estimator=clf,param_grid=params)
gcv.fit(X_train,y_train)
```

```python
print(f"Best Hyperparameters: {gcv.best_params_}")
```

```python
model = gcv.best_estimator_
print(gcv.best_estimator_)
model.fit(X_train,y_train)
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)


# Calculate and print the R-squared scores
print(f'Train score{r2_score(y_train,y_train_pred)}')
print(f'Test score{r2_score(y_test,y_test_pred)}')

```
